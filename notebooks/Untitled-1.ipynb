```python
import numpy as np
import matplotlib.pyplot as plt


def get_positional_encoding(max_sequence_len, d_model):
    assert d_model % 2 == 0, "d_model must be even for sinusoidal encoding."
    pos_enc = np.zeros((max_sequence_len, d_model))

    # add additional dimension, place array around each index
    position = np.arange(max_sequence_len)[:, np.newaxis]
    # from positional encoding equation (remeber that exp(aln(b)) = b ** a)
    div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))

    pos_enc[:, 0::2] = np.sin(position * div_term)
    pos_enc[:, 1::2] = np.cos(position * div_term)

    return pos_enc
```

```python
LENGTH = 100
DIM_MODEL = 512
pe_matrix = get_positional_encoding(LENGTH, DIM_MODEL)

plt.figure(figsize=(10, 8))
plt.imshow(pe_matrix, cmap='RdBu', aspect='auto')
plt.title("Positional Encoding Verification")
plt.xlabel("Dimension (d_model)")
plt.ylabel("Token Position")
plt.show()
```
